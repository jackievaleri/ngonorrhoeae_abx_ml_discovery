```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as sp
import seaborn as sns
```

```python
newdf = pd.DataFrame(columns=["Value", "Metric", "Model"])

labels = ["RFC", "SVM", "FFN", "GNN"]

for fold, name in zip(
    ["rfc_final_122", "svm_final_15", "ffn_final_20", "FINALbayHO04052022"], labels
):

    if "GNN" not in name:
        df = pd.read_csv("../models/" + fold + "/test_scores.csv")
        new = pd.DataFrame()
        new["Value"] = list(df.iloc[0, 3:33])  # 30 folds not 20
        new["Metric"] = ["auPR"] * len(new)
        new["Model"] = name
        newdf = newdf.append(new)

        new = pd.DataFrame()
        new["Value"] = list(df.iloc[0, 35:65])  # 30 folds not 20
        new["Metric"] = ["auROC"] * len(new)
        new["Model"] = name
        newdf = newdf.append(new)
    else:
        df = pd.read_csv("../models/" + fold + "/test_scores.csv")
        new = pd.DataFrame()
        new["Value"] = list(df.iloc[0, 3:8])  # 30 folds not 20
        new["Metric"] = ["auPR"] * len(new)
        new["Model"] = name
        newdf = newdf.append(new)

        new = pd.DataFrame()
        new["Value"] = list(df.iloc[0, 10:15])  # 30 folds not 20
        new["Metric"] = ["auROC"] * len(new)
        new["Model"] = name
        newdf = newdf.append(new)

df = newdf
# df = df.set_index('Task')
df
```

```python
# manually add in chemberta results from zinc pretraining + 77M pubchem pretraining
# auPRC: 0.470, 0.425
# auROC: 0.913, 0.899
newrows = pd.DataFrame()
newrows["Value"] = [0.470, 0.425, 0.913, 0.899]
newrows["Metric"] = ["auPR", "auPR", "auROC", "auROC"]
newrows["Model"] = ["ChemBERTa"] * len(newrows)

df = df.append(newrows)
df
```

```python
# new order so GNN is last
from pandas.api.types import CategoricalDtype

cat_model_order = CategoricalDtype(
    ["RFC", "SVM", "FFN", "ChemBERTa", "GNN"], ordered=True
)
df["Model"] = df["Model"].astype(cat_model_order)
df.sort_values("Model")
df
```

```python
plt.figure(figsize=(5, 3), dpi=300)
my_pal = dict(
    zip(
        ["RFC", "SVM", "FFN", "ChemBERTa", "GNN"],
        ["lightgrey", "darkgrey", "slategrey", "dimgrey", "royalblue"],
    )
)
ax = sns.barplot(
    data=df,
    x="Metric",
    y="Value",
    hue="Model",
    edgecolor="black",
    linewidth=1,
    alpha=0.7,
    palette=my_pal,
)
sns.swarmplot(
    data=df,
    x="Metric",
    y="Value",
    hue="Model",
    dodge=True,
    edgecolor="black",
    linewidth=0.2,
    s=2,
    label="",
    palette=my_pal,
)
handles, labs = ax.get_legend_handles_labels()
ax.legend(
    handles[:5], labs[:5], title="Model", bbox_to_anchor=(1.01, 1), borderaxespad=0
)
plt.xlabel("Metric")
plt.ylim([0, 1.0])
plt.xticks(rotation=60)
plt.ylabel("Area Under Curve")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()

plt.savefig("../out/all_model_comparison_10132022.svg")
plt.savefig("../out/all_model_comparison_10132022.png")
plt.show()
```

```python
gnns = df[df["Model"] == "GNN"]
gnn_aurocs = list(gnns[gnns["Metric"] == "auROC"]["Value"])
gnn_auprs = list(gnns[gnns["Metric"] == "auPR"]["Value"])

ffns = df[df["Model"] == "FFN"]
ffn_aurocs = list(ffns[ffns["Metric"] == "auROC"]["Value"])
ffn_auprs = list(ffns[ffns["Metric"] == "auPR"]["Value"])

svms = df[df["Model"] == "SVM"]
svm_aurocs = list(svms[svms["Metric"] == "auROC"]["Value"])
svm_auprs = list(svms[svms["Metric"] == "auPR"]["Value"])

rfcs = df[df["Model"] == "RFC"]
rfc_aurocs = list(rfcs[rfcs["Metric"] == "auROC"]["Value"])
rfc_auprs = list(rfcs[rfcs["Metric"] == "auPR"]["Value"])

chembertas = df[df["Model"] == "ChemBERTa"]
chemberta_aurocs = list(chembertas[chembertas["Metric"] == "auROC"]["Value"])
chemberta_auprs = list(chembertas[chembertas["Metric"] == "auPR"]["Value"])
```

```python
display(sp.ttest_ind(gnn_aurocs, ffn_aurocs))
display(sp.ttest_ind(gnn_aurocs, svm_aurocs))
display(sp.ttest_ind(gnn_aurocs, rfc_aurocs))
display(sp.ttest_ind(gnn_aurocs, chemberta_aurocs))
```

```python
display(sp.ttest_ind(gnn_auprs, ffn_auprs))
display(sp.ttest_ind(gnn_auprs, svm_auprs))
display(sp.ttest_ind(gnn_auprs, rfc_auprs))
display(sp.ttest_ind(gnn_auprs, chemberta_auprs))
```
